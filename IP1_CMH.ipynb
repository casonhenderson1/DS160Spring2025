```python
car_sales['Price']=car_sales['Price'].astype(str).str.replace('$', '')
car_sales['Price']=car_sales['Price'].astype(str).str.replace(',', '').astype('int64') 
car_sales.head()
```

```python
car_sales['Total Sales'] = car_sales['Price'].cumsum()
car_sales.head()
```

```python
car_sales['Sale Date']=pd.date_range('02/9/2022', periods=len(car_sales))
car_sales.head()
```

```python
plt.plot(car_sales['Price'], car_sales['Total Sales'])
plt.title('Price vs Total Sales')
plt.xlabel('Price')
plt.ylabel('Total Sales')
plt.show()
```

```python
plt.scatter(car_sales['Odometer (KM)'], car_sales['Price'])
plt.title('Odometer (KM) vs Price')
plt.xlabel('Odometer (KM)')
plt.ylabel('Price')
plt.show()
```

```python
avg_odometer=car_sales.groupby('Make')['Odometer (KM)'].mean()
avg_odometer.plot(kind='bar')
plt.title('Average Odometer (KM) by Make')
plt.xlabel('Make')
plt.ylabel('Average Odometer (KM)')
plt.xticks(rotation=360)
plt.show()
```

```python
make_counts=car_sales['Make'].value_counts()
make_counts.plot(kind='bar')
plt.title('Histogram of Car Makes')
plt.xlabel('Make')
plt.ylabel('Frequency')
plt.xticks(rotation=360)
plt.show()
```

```python
fig, axes=plt.subplots(nrows=2, ncols=2,figsize=(14, 10))
axes[0, 0].plot(car_sales['Price'], car_sales['Total Sales'])
axes[0, 0].set_title('Price vs Total Sales')
axes[0, 0].set_xlabel('Price')
axes[0, 0].set_ylabel('Total Sales')

axes[0, 1].scatter(car_sales['Odometer (KM)'], car_sales['Price'])
axes[0, 1].set_title('Odometer (KM) vs Price')
axes[0, 1].set_xlabel('Odometer (KM)')
axes[0, 1].set_ylabel('Price')

average_odometer=car_sales.groupby('Make')['Odometer (KM)'].mean()
average_odometer.plot(kind='bar', ax=axes[1, 0])
axes[1, 0].set_title('Average Odometer (KM) by Make')
axes[1, 0].set_xlabel('Make')
axes[1, 0].set_ylabel('Average Odometer (KM)')
axes[1, 0].tick_params(axis='x', rotation=360)

make_counts=car_sales['Make'].value_counts()
make_counts.plot(kind='bar')
axes[1, 1].set_title('Histogram of Car Makes')
axes[1, 1].set_xlabel('Make')
axes[1, 1].set_ylabel('Frequency')
axes[1, 1].tick_params(axis='x', rotation=360)
```

```python
import seaborn as sns
sns.boxplot(x='Colour', y='Total Sales', data=car_sales)
plt.title('Box Plot of Colour vs Total Sales')
plt.xlabel('Colour')
plt.ylabel('Total Sales')
plt.show()
```

```python
sns.lmplot(x='Odometer (KM)', y='Total Sales', data=car_sales, hue='Make', ci=None, height=6)
plt.title('Linear Regression of Total Sales vs Odometer (KM)')
plt.xlabel('Odometer (KM)')
plt.ylabel('Total Sales')
plt.show()
```

```python
sns.pairplot(car_sales)
plt.suptitle('Pair Plot for All Possible Combinations', y=1.02)
plt.show()
```

```python
correlation = car_sales[['Total Sales', 'Price', 'Odometer (KM)']].corr()
correlation
```

```python
sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0, linewidths=0.5, linecolor='black')
plt.title('Pearson Correlation Heatmap')
plt.show()
```

For Total Sales and Price - A high positive correlation of 1 means that as the price increases, the total sales tend to increase as well.
For the Total Sales and Odometer (KM) - The weak negative correlation of -0.32 alludes to a slight tendency that as the odometer reading increases, the total sales decrease. 
For the Price and Odometer (KM) - A weak negative correlation 0.16 suggests a slight tendency that as the odometer reading increases, the price decreases.
