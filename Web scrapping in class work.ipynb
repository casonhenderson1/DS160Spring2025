```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://en.wikipedia.org/wiki/List_of_Super_Bowl_champions'
response = requests.get(url)
```

```python

# Check if the request was successful
if response.status_code == 200:
    print("Successfully fetched the webpage!")
else:
    print("Failed to fetch the webpage.")
```

```python

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

tables=soup.find_all('table')
table_sbchamp=tables[1]

if table_sbchamp:
    print('table_found')

df_sbchamp=pd.read_html(str(table_sbchamp))[0]
df_sbchamp.head()
```

### Task 3: Extract Elements

```python
#Find all images and extract their src attributes.
images = soup.find_all('img')
image_urls = [img['src'] for img in images if 'src' in img.attrs]
```

```python
# Locate and extract all tables on the webpage, converting them into Pandas DataFrames.
import pandas as pd

tables = soup.find_all('table')
for i, table in enumerate(tables):
    df = pd.read_html(str(table))[0]
    print(f"Table {i+1}:\n", df.head(), "\n")
```
